### Barplot

```python
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

# load data_set
#chilla=pd.read_csv("data_viz.csv")
chilla=sns.load_dataset("iris")

#plot with 2 varriables
p=sns.barplot(x="species",y="petal_width", data=chilla)
p.set_title("Bar plot for Chilla participant age w.r.t with 2nd Varriable")
plt.show()
```

```python
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

# load data_set
#chilla=pd.read_csv("data_viz.csv")
chilla=sns.load_dataset("titanic")

#plot with 2 varriables
p=sns.barplot(x="who",y="alone", hue="sex",data=chilla,order=["woman","man"],color="grey",ci=None,
             palette='pastel')
p.set_title("Bar plot for Chilla participant age w.r.t with 2nd Varriable")
plt.show()
```

```python
import pandas as pd
from numpy import mean
import seaborn as sns
import matplotlib.pyplot as plt

# load data_set
#chilla=pd.read_csv("data_viz.csv")
chilla=sns.load_dataset("titanic")

#plot with 2 varriables
p=sns.barplot(x="class",y="fare", hue="sex",data=chilla,color="grey",estimator=mean)
p.set_title("Bar plot for Chilla participant age w.r.t with 2nd Varriable")
plt.show()
```
