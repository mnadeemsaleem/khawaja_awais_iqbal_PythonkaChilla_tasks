### How to mange individual color to hue color

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="whitegrid")
tips = sns.load_dataset("tips")
# using pelette aurument we manage hue color(solution)
pelt={"Yes": "indigo", "No": "green"}
#sns.boxplot(x="day", y="tip", data=tips,hue='smoker',saturation=.5,palette=pelt)

# Another way 
sns.boxplot(x="day", y="tip", data=tips,hue='smoker',saturation=.5,
            palette={"Yes": "red", "No": "green"})
       
plt.title("Daywise tip from smoker")
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
kashti=sns.load_dataset("titanic")
sns.boxplot(x="survived",y="age",data=kashti,showmeans=True,
           meanprops={"marker":"*","markersize":"12",
                       "markeredgecolor":"yellow"})
# show labels
plt.xlabel("How many servived",size=12),
plt.ylabel("Age (years)",size=12),
plt.title("Boxplot for kinay dobay or kinay bacha",size=18,weight="bold")

plt.show()
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

nuqta=sns.load_dataset("dots")
nuqta.head()
#defining a color pelette
#pel= sns.color_palette('rocket_r')

```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

nuqta=sns.load_dataset("dots")

#defining a color pelette
pel= sns.color_palette('rocket_r')

# creating plot
sns.set_theme(style="ticks")

dots = sns.load_dataset("dots")

# Define the palette as a list to specify exact values
palette = sns.color_palette("rocket_r")

# Plot the lines on two facets
sns.relplot(
    data=dots,
    x="time", y="firing_rate",
    hue="coherence", size="choice", col="align",
    kind="line", size_order=["T1", "T2"], palette=pel,
    height=5, aspect=.75, facet_kws=dict(sharex=False),
)

plt.title("Nuqta ka plot")
plt.show()
```
